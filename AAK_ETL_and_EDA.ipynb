Let's handle this missing values for. engv. That is a numerical data. We will be filling the missing values with medium. As medium is more robust to outliers and also preserves the distribution of the data without introducing any bias. As for drive that is categorical, we will be assigning unknown We will be just assigning a node. So prevented data loss by keeping rows with missing values. And it will also allow the model to treat unknown as distinct category.

```python
#handling missing values
#for Numerical data
car_sales_data['engV'].fillna(car_sales_data['engV'].median(), inplace=True)
#for Categorical Data
car_sales_data['drive'].fillna('unknown', inplace=True)
```

```python
#checking missing values after handling them
car_sales_data.isnull().sum()
```

```python
# Handling Duplicates : If there are any duplicates we will be romving them
car_sales_data.duplicated().sum()
```

```python
#removing duplicates
car_sales_data.drop_duplicates(inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Step 4: Distribution Analysis

# Price Distribution
plt.figure(figsize=(8, 5))
sns.histplot(car_sales_data['price'], bins=50, kde=True, color='blue')
plt.title('Price Distribution')
plt.xlabel('Price (USD)')
plt.ylabel('Frequency')
plt.show()

# Mileage Distribution
plt.figure(figsize=(8, 5))
sns.histplot(car_sales_data['mileage'], bins=50, kde=True, color='green')
plt.title('Mileage Distribution')
plt.xlabel('Mileage (1000 km)')
plt.ylabel('Frequency')
plt.show()

# Engine Volume Distribution
plt.figure(figsize=(8, 5))
sns.histplot(car_sales_data['engV'], bins=30, kde=True, color='purple')
plt.title('Engine Volume Distribution')
plt.xlabel('Engine Volume (L)')
plt.ylabel('Frequency')
plt.show()

# Step 5: Relationship Analysis

# Price vs. Mileage
plt.figure(figsize=(8, 5))
sns.scatterplot(x='mileage', y='price', data=car_sales_data, hue='body', alpha=0.7)
plt.title('Price vs Mileage by Body Type')
plt.xlabel('Mileage (1000 km)')
plt.ylabel('Price (USD)')
plt.legend(title='Body Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Price vs Year
plt.figure(figsize=(8, 5))
sns.boxplot(x='year', y='price', data=car_sales_data, showfliers=False)
plt.title('Price vs Year')
plt.xlabel('Year')
plt.ylabel('Price (USD)')
plt.xticks(rotation=45)
plt.show()

# Step 6: Categorical Analysis

# Average price by body type
avg_price_by_body = car_sales_data.groupby('body')['price'].mean().sort_values()

plt.figure(figsize=(8, 5))
avg_price_by_body.plot(kind='bar', color='skyblue')
plt.title('Average Price by Body Type')
plt.xlabel('Body Type')
plt.ylabel('Average Price (USD)')
plt.xticks(rotation=45)
plt.show()

# Average price by engine type
avg_price_by_engType = car_sales_data.groupby('engType')['price'].mean().sort_values()

plt.figure(figsize=(8, 5))
avg_price_by_engType.plot(kind='bar', color='orange')
plt.title('Average Price by Engine Type')
plt.xlabel('Engine Type')
plt.ylabel('Average Price (USD)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Step 7: Price Trends by Car Brands

# Average price by car brand
avg_price_by_brand = car_sales_data.groupby('car')['price'].mean().sort_values(ascending=False)

# Visualize the top 15 brands by average price
top_15_brands = avg_price_by_brand.head(15)

plt.figure(figsize=(10, 6))
top_15_brands.plot(kind='bar', color='teal')
plt.title('Top 15 Car Brands by Average Price')
plt.xlabel('Car Brand')
plt.ylabel('Average Price (USD)')
plt.xticks(rotation=45)
plt.show()

# Count of cars by brand (popularity)
brand_counts = car_sales_data['car'].value_counts().head(15)

plt.figure(figsize=(10, 6))
brand_counts.plot(kind='bar', color='coral')
plt.title('Top 15 Car Brands by Count')
plt.xlabel('Car Brand')
plt.ylabel('Number of Cars')
plt.xticks(rotation=45)
plt.show()
```

Luxury brands such as Bentley and Ferrari dominate the higher end of the average price spectrum and premium brands like Mercedes Benz, also have a high average price. Popular brands in the data set include Volkswagen, Mercedes Benz and BMW in Toyota, which. are more affordable and widely available.

```python
# Combine brand popularity (count) with average price for comparison
brand_popularity_price = car_sales_data.groupby('car').agg(
    avg_price=('price', 'mean'),
    count=('car', 'size')
).sort_values(by='count', ascending=False)

# Extract the top 15 brands by count for comparison
top_15_popular_brands = brand_popularity_price.head(15)

# Plotting average price and count side by side
fig, ax1 = plt.subplots(figsize=(12, 6))

# Bar plot for car counts (popularity)
ax1.bar(top_15_popular_brands.index, top_15_popular_brands['count'], color='coral', label='Count', alpha=0.7)
ax1.set_xlabel('Car Brand')
ax1.set_ylabel('Count', color='coral')
ax1.tick_params(axis='y', labelcolor='coral')
ax1.set_xticklabels(top_15_popular_brands.index, rotation=45)

# Line plot for average price
ax2 = ax1.twinx()
ax2.plot(top_15_popular_brands.index, top_15_popular_brands['avg_price'], color='blue', marker='o', label='Avg Price')
ax2.set_ylabel('Average Price (USD)', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# Title and legends
fig.suptitle('Comparison of Brand Popularity and Average Price')
fig.tight_layout()
plt.show()
```

Popular brands like Volkswagen Toyota and Mercedes have large number of cars, but moderate average prices reflecting their. broader availability and affordability. Luxury plans like. Lexus. BMW Audi. show lower counts, but higher average prices indicating their niche market appeared. There is also an inverse trend for some brands where high popularity correlates with lower prices. Example Toyota while brands with low counts. exhibit significantly higher prices.

```python
#price trend accross manufacturing Years

avg_car_price_by_year = car_sales_data.groupby('year')['price'].mean()

plt.figure(figsize=(10, 6))
avg_car_price_by_year.plot(kind='line', marker='o', color='green')
plt.title('Average Car Price by Manufacturing Year')
plt.xlabel('Year')
plt.ylabel('Average Price (USD)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

#calculate the number of cars per year

car_count_by_year = car_sales_data.groupby('year').value_counts().sort_index()

plt.figure(figsize=(10, 6))
car_count_by_year.plot(kind='bar', color='skyblue')
plt.title('Number of  Car  by Manufacturing Year')
plt.xlabel('Year')
plt.ylabel('Number of  Car')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Car manufacture in the recent years have higher average prices, reflecting that near models and are equipped with better technology. And there is also a noticeable decline in average price for older cars, likely due to depreciation.
